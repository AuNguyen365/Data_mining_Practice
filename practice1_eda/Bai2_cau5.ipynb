Ý nghĩa của chuẩn hóa dữ liệu trong học máy

Chuẩn hóa dữ liệu (Normalization) là quá trình đưa các giá trị của các đặc trưng về cùng một thang đo, thường trong khoảng 0 đến 1 hoặc -1 đến 1.

#### **1. Giúp các đặc trưng có cùng thang đo**

Trong dataset, các đặc trưng có thể có đơn vị và giá trị rất khác nhau.

ví dụ:
| Feature | Giá trị |
|--------|--------|
| chiều cao | 170 |
| cân nặng | 60 |
| tuổi | 20 |

Nếu không chuẩn hóa thì đặc trưng có giá trị lớn sẽ ảnh hưởng mạnh hơn đến mô hình.
Chuẩn hóa giúp đưa tất cả các đặc trưng về cùng khoảng giá trị, giúp mô hình học công bằng hơn.

#### **2. Cải thiện hiệu suất của thuật toán học máy**

Nhiều thuật toán học máy hoạt động dựa trên khoảng cách giữa các điểm dữ liệu, ví dụ:
K-Nearest Neighbors (KNN)
K-Means Clustering
Support Vector Machine (SVM)

Nếu dữ liệu không chuẩn hóa thì đặc trưng có giá trị lớn sẽ chi phối việc tính khoảng cách.
Chuẩn hóa giúp thuật toán đánh giá đúng mức độ quan trọng của từng đặc trưng.

#### **3. Giúp mô hình học nhanh và ổn định hơn**

Chuẩn hóa dữ liệu giúp:

Tăng tốc độ hội tụ của thuật toán
Giảm sai số trong quá trình huấn luyện

Điều này đặc biệt quan trọng với các mô hình như:
Gradient Descent
Neural Network

#### **4. Ví dụ chuẩn hóa bằng Min-Max Scaling**

Trong Python có thể chuẩn hóa dữ liệu bằng thư viện scikit-learn:

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler

# Load dataset
iris = load_iris()

# Tạo dataframe
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Chuẩn hóa dữ liệu
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)

# Hiển thị 5 dòng đầu
print(normalized_data[:5])
```
