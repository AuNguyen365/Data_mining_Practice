Khai báo thư viện

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

1. Nạp dữ liệu và hiển thị 5 dòng đầu tiên.

```python
df = pd.read_csv('train.csv') # nạp dữ liệu

print(df.head()) # hiển thị 5 dòng đầu
```

2. Xác định có bao nhiêu dòng, bao nhiêu cột.

```python
print("\n--- Kích thước tập dữ liệu ---")
print(f"Số dòng: {df.shape[0]}")
print(f"Số cột: {df.shape[1]}")
```

3. Thống kê các cột dạng số (describe()), đếm giá trị thiếu.

```python
print(df.describe()) # thống kê
```

```python
print("\nSố lượng giá trị thiếu ở mỗi cột:")
print(df.isnull().sum()) # đếm số ô trống 
```

4. Vẽ histogram cho thuộc tính Age và boxplot cho Fare.

```python
plt.figure(figsize=(12, 5))
```

Histogram cho Age

```python
# Histogram cho Age
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True, color='blue')
plt.title('Phân bố tuổi tác (Age)')
plt.show()
```

Boxplot cho Fare

```python
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'], color='red')
plt.title('Biểu đồ hộp cho giá vé (Fare)')
plt.show()
```

5. Điền giá trị trung vị cho Age bị thiếu. 

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

print(f"Giá trị trung vị của Age là: {median_age}")
print(f"Số giá trị thiếu của Age sau khi xử lý: {df['Age'].isnull().sum()}")
```

6. Chuẩn hóa các cột Age và Fare. 


```python
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("\n--- Dữ liệu Age và Fare sau khi chuẩn hóa (hiển thị 5 dòng) ---")
display(df[['Age', 'Fare']].head())
```

7. Mã hóa cột Sex và Embarked thành dạng số. 

```python
# Mã hóa Sex (male/female -> 0/1)
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])

# Mã hóa Embarked (Xử lý thiếu trước khi mã hóa)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked'] = le.fit_transform(df['Embarked'])

print("Dữ liệu sau khi chuẩn hóa và mã hóa:")
display(df[['Age', 'Fare', 'Sex', 'Embarked']].head())
```

8. Lưu tập dữ liệu sau xử lý thành titanic_clean.csv. 

```python
df.to_csv('titanic_clean.csv', index=False)
print("Đã lưu file titanic_clean.csv thành công!")
```
